## Investigation Overview

> This exploration and visualization study mainly focuses on finding the bike usage pattern and customer habit or routine, specifically how user type, time, gender, and age impacts usage of the system.

## Dataset Overview

> This data set includes information about individual rides made in a bike-sharing system covering the greater San Francisco Bay area during february. Visualizations below are created from wrangled and cleaned data to facilitate exploration analysis and help discover usage pattern and rider characteristics jumping to conclusions to their behaviors.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv('Fordgobike_Exploration/cleaned_fordgobike.csv')
```

## Average Trip duration in Weekdays for Different user Types

>- Subscribers tend to have shorter trip duration than customers
>- They both increase a little in weekend holidays

```python
weekday = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
weekdaycat = pd.api.types.CategoricalDtype(ordered=True, categories=weekday)
df['start_day_of_week'] = df['start_day_of_week'].astype(weekdaycat)


plt.figure(figsize=[8, 6]);
sb.pointplot(data=df, x='start_day_of_week', y='duration_min', hue='user_type', dodge=0.3, linestyles="");
plt.legend(title='User Type');
plt.xlabel('Day of Week',fontsize= 12);
plt.ylabel('Average Trip Duration (minutes)',fontsize= 12);
plt.suptitle('Average Weekday Trip Duration for Customers and Subscribers', fontsize=14, fontweight='semibold', y=1.015);
```

## Trip Duration for Different user Types and Genders
>- Females take longer Time duration than males and others weather in both user types


```python
graph = sb.catplot(data=df, x='user_type', y="duration_min", hue="member_gender", kind="violin", height=5, aspect=2);

graph.set_axis_labels("User Type", "Trip Duration (Min)")
graph._legend.set_title('Gender')
graph.fig.suptitle('Trip Duration Per User Type and Gender', y=1.05, fontsize=16, fontweight='bold');
```

## Trip Counts Per week day for all Genders

> All Genders use Bikes for the same purpose which is in wokring days since all genders have the same wave, however the number of men are extremely higher than women and therfore offers can be made to tackle such a gap in gender

```python
lm = sb.catplot(data=df, x='start_day_of_week', hue='member_gender', kind='count',aspect=20.7/8.27, palette='viridis_r')

lm.set_axis_labels('Day', 'Rides Count')
lm._legend.set_title('Gender')
lm.fig.suptitle('Overall Trips Usage of Bike Rides Per Gender', fontsize=14,fontweight='semibold');
```
